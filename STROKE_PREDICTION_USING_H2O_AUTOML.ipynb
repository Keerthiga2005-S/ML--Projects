```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
d=h2o.import_file("/content/healthcare-dataset-stroke-data.csv")
d.head()
```

```python
d.isna().sum()
```

```python
d.shape
```

```python
d=d.drop("bmi",axis=1)
```

```python
d=d.drop('gender',axis=1)
```

```python
d=d.drop('ever_married',axis=1)
```

```python
d=d.drop('id',axis=1)
```

```python
d.isna().sum()
```

```python
x=d.columns[:-1]
y=d.columns[-1]
d[y]=d[y].asfactor()

train,test=d.split_frame(ratios=[0.8])
```

```python
aml=H2OAutoML(max_models=10,seed=1)
aml.train(x=x,y=y,training_frame=train)
```

```python
lb=aml.leaderboard
print(lb)
```

```python
best_model=aml.leader
predictions=best_model.predict(test)
```

```python
performance=best_model.model_performance(test)
print(performance)
```
