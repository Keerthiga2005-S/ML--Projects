```python
def gradient_descent(x,y):
  m_curr=c_curr=0
  iterations=1000
  learning_rate=0.08
  n=len(x)
  y_p=[]
  for i in range(iterations):
    y_predicted=m_curr*x+c_curr   #y=mx+c
    y_p.append(y_predicted)
    cost=(1/n)*sum([val**2 for val in (y-y_predicted)])
    md=-(2/n)*sum(x*(y-y_predicted))
    cd=-(2/n)*sum(y-y_predicted)
    m_curr=m_curr-learning_rate*md
    c_curr=c_curr-learning_rate*cd
    print('m {},c {},cost {},iterations {}'.format(m_curr,c_curr,cost,i))
  return y_p
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
y_p=gradient_descent(x,y)

```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.title("LINEAR REGRESSION I226-KEERTHIGA S")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
for i in y_p:
  plt.plot(x,i)
```
